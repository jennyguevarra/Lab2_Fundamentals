```python
target_name = "GUEVARRA"
SEED_NUM = 2
redacted_name = ""

for ch in target_name:
    if ch in "AEIOU":
        redacted_name += str(SEED_NUM)
    else:
        redacted_name += ch

print("Original Surname:", target_name)
print("SEED_NUM Used:", SEED_NUM)
print("Redacted Output:", redacted_name)
```

```python
ENGINEER = "GUEVARRA"   
SEED_NUM = 2            
loops = 7               

A = ord(ENGINEER[0])    
B = SEED_NUM
C = loops

verification_hash = (A * B) - C

print("First Letter ASCII (A):", A)
print("SEED_NUM (B):", B)
print("Total Loops (C):", C)
print("Verification Hash:", verification_hash)
```

```python
correct_password = "SECRET"
attempts = 0

while True:
    pwd = input("Enter Password: ")
    attempts += 1
    if pwd == correct_password:
        print("Access Granted.")
        print("Attempts Before Success:", attempts)
        break
    else:
        print("Access Denied.")
```
